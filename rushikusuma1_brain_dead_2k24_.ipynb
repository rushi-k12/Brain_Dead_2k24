```python
import pandas as pd
import numpy as np
import zipfile
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, concatenate, Dropout
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from PIL import Image

# Step 1: Extract the Image Data
with zipfile.ZipFile("/content/drive/MyDrive/images.zip", 'r') as zip_ref:
    zip_ref.extractall("/content/images")

# Step 2: Load and Preprocess the Data
# Load CSV file containing text data into a DataFrame
text_data = pd.read_csv("/content/haemful_meme_for_covid19.csv")

# Drop any rows with NaN values in the 'text' column
text_data.dropna(subset=['text'], inplace=True)

# Preprocess text and image data as required
# For text data: tokenize, pad sequences
# For image data: resize, normalize, etc.

# Step 3: Prepare the Data
# Combine text and image data into suitable formats for training
# For example, convert text data into sequences and preprocess images

# Text data preprocessing
max_words = 10000  # Maximum number of words in the vocabulary
max_sequence_length = 100  # Maximum length of input sequences
tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(text_data['text'].astype(str))  # Convert to string type before fitting tokenizer
sequences = tokenizer.texts_to_sequences(text_data['text'].astype(str))
text_input_data = pad_sequences(sequences, maxlen=max_sequence_length)

# Image data preprocessing (using pretrained ResNet50 model)
image_input_data = []  # Store extracted image features
for meme_id in text_data['image']:
    image_path = os.path.join("/content/images", meme_id)
    image = Image.open(image_path)
    image = image.resize((224, 224))  # Resize image to match ResNet50 input size
    image = np.array(image) / 255.0  # Normalize pixel values
    image_input_data.append(image)

image_input_data = np.array(image_input_data)

# Convert class labels to numpy array
labels = np.array([1 if label == 'harmful' else 0 for label in text_data['labels']])

# Step 4: Define and Compile the Model
# Define model architecture
text_input = Input(shape=(max_sequence_length,), name='text_input')
image_input = Input(shape=(224, 224, 3), name='image_input')
text_output = Dense(128, activation='relu')(text_input)
image_model = ResNet50(weights='imagenet', include_top=False)
for layer in image_model.layers:
    layer.trainable = False
image_output = image_model(image_input)
image_output = Dense(128, activation='relu')(image_output)
fusion_layer = concatenate([text_output, image_output])
fusion_layer = Dropout(0.5)(fusion_layer)
output = Dense(1, activation='sigmoid')(fusion_layer)

# Create and compile the model
model = Model(inputs=[text_input, image_input], outputs=output)
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Step 5: Train the Model
X_train_text, X_test_text, X_train_image, X_test_image, y_train, y_test = train_test_split(
    text_input_data, image_input_data, labels, test_size=0.2, random_state=42
)

model.fit([X_train_text, X_train_image], y_train, epochs=10, batch_size=32, validation_split=0.1)

# Step 6: Evaluate the Model
y_pred = model.predict([X_test_text, X_test_image])
roc_auc = roc_auc_score(y_test, y_pred)
print("ROC AUC Score:", roc_auc)
```
